```python
import matplotlib.pyplot as plt
import yfinance as yf

alpha_vantage_api_key  = 'S93OP37YPB6MPLYF'
data = yf.download('GC=F', start='2023-01-01', end='2023-07-16')

def plot_gold_data(data, key):

    # Plotting the closing price
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data[key], color='gold')
    plt.title('Latest GOLD Data')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

plot_gold_data(data=data, key='Close')
```
